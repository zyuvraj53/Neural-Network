```python
# We're going to have a target column in our weather data to store the predictions, it will initially have tomorrows temperature
weather["target"] = weather.shift(-1)["tmax"] # shift helps us do that
```

```python
from sklearn.linear_model import Ridge
```

```python
weather.corr()
```

```python
rr = Ridge(alpha=.1)
```

```python
# We will create a predictors dataset, which will help us predict tomorrow's weather
predictors = weather.columns[~weather.columns.isin(["target", "name", "station"])] # This will give us all the columns which
                                                                                   # are not in the list
```

```python
predictors
```

```python
# This is time-series data, whenever we want to calculate the error of a machine learning model, we can use cross-validation
# except with time-series data !!! We need to be careful not to use future data to predict the past
## We will be using backtesting
def backtest(weather=None, model=None , predictors=None, start=3650, step=90): # We want 10 years of predictions, and every 90 days, we're 
                                                                # going to predict the next day's data
        all_predictions = []
        
        for i in range(start, weather.shape[0], step):
            train = weather.iloc[:i,:]
            test = weather.iloc[i:(i+step),:]
            
            model.fit(train[predictors], train["target"])
            
            preds = model.predict(test[predictors])
            
            preds = pd.Series(preds, index=test.index)
            combined = pd.concat([test["target"], preds], axis=1) # axis = 1, means that everything that we passed in is to be treated as a separate column
            
            combined.columns = ["actual", "predictions"]
            
            combined["diff"] = (combined["prediction"] - combined["actual"]).abs()
            
            all_predictions.append(combined)
        return pd.concat(all_predictions) # by default axis = 0, means treat everything as separate rows
```

```python
predictions = backtest(weather, rr, predictors)
```
