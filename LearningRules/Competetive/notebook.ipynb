```python
import numpy as np
import random
import matplotlib.pyplot as plt
```

#### In competetive learning, there are two sets of connections,
#### 1) the connections between input neurons and output neurons, called feedforward or excitatory connections
#### 2) the connections in-between output neurons, called lateral or inhibitory connections

##### There should be altleat 2 output neurons

##### for a neuron k
##### y_k = 1 if v_k > v_j for all j, j != k
##### y_k = 0 otherwise
##### sum_over_j w_kj = 1 for all k
##### sum_over_j w_kj^2 = 1 for all k

```python

# Generate some random 2D data points
np.random.seed(42)
data = np.concatenate([np.random.randn(50, 2) + np.array([2, 2]),
                      np.random.randn(50, 2) + np.array([-2, -2])])

plt.scatter(data[:, 0], data[:, 1], label='Data Points')
plt.legend()
plt.title('Plotting points')
plt.show()
```

```python

weights = np.random.rand(2, 2)
weights
```

```python

learning_rate = 0.1
epochs = 100
```

```python

for epoch in range(epochs):
    for point in data:
        # Calculate the Euclidean distance between the point and each neuron's weight
        distances = np.linalg.norm(weights - point, axis=1)
        print(f"distance: {distances}")
        
        winner = np.argmin(distances)
        print(f"winner: {winner}")
        
        weights[winner] += learning_rate * (point - weights[winner])
        print()
```

```python

# Plot the results
plt.scatter(data[:, 0], data[:, 1], label='Data Points')
plt.scatter(weights[:, 0], weights[:, 1], c='red', label='Neurons', marker='X')
plt.legend()
plt.title('Competitive Learning: Clustering 2D Points')
plt.show()
```
